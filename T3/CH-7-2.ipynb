```python
def forecast():
    api_key = "011dd5500f9b18f250cfafe2ff91cad5"
    city = input("Enter City Name :")
    
    req_url = f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    res = requests.get(req_url)
    coords = res.json()
    # print(json.dumps(coords,indent=5))
    lat = coords[0]['lat']
    lon = coords[0]['lon']
    
    req_url2 = f'http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}'
    res2 = requests.get(req_url2)
    weather_data1 = res2.json()
    print(json.dumps(weather_data1,indent=5))
    return weather_data1
wd1 = forecast()
```
